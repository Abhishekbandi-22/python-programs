```python
def attendance_management():
    attend = []
    for i in range(5):
        attende = input("Enter attendance (P/A): ")
        if attende == "P" or attende == "p":
            attend.append("P")
        else:
            attend.append("A")
    print("Attendance Record:", attend)
    pc = attend.count("P")
    if pc >= 5:
        print("Regular")
    elif pc == 3:
        print("Warning")
    else:
        print("Irregular")

attendance_management()
```

```python
def calculate_salary():
    basic_salary = float(input("Enter basic salary: "))
    present_days = int(input("Enter present days: "))

    HRA = basic_salary * 0.20
    DA = basic_salary * 0.10
    PF = basic_salary * 0.12

    net_salary = basic_salary + HRA + DA - PF
    if present_days < 3:
        print("Salary is not credited due to low attendance")
    else:
        print("Net Salary:", net_salary)

calculate_salary()
```

```python
def performance():
    ratings = []
    print("Enter 3 Performance Ratings (1–5):")
    for i in range(3):
        r = int(input("Rating: "))
        ratings.append(r)
    avg_rating = sum(ratings) / 3
    print("Average Rating:", avg_rating)
    if avg_rating >= 4:
        print("Performance: Excellent")
    elif avg_rating >= 3:
        print("Performance: Good")
    else:
        print("Performance: Needs Improvement")

performance()
```
