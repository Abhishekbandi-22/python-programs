```python
print("=== Digital Dreams Course Fee Billing System ===")
print("\nAvailable Courses:")
print("1. Python Programming - ₹5000")
print("2. Data Analytics - ₹8000")
print("3. AI & ML - ₹12000")

course_choice = int(input("\nEnter course number (1-3): "))
is_student = input("Are you a student? (yes/no): ").lower()
early_registration = input("Early registration? (yes/no): ").lower()


course_name = ""
original_fee = 0


if course_choice == 1:
    course_name = "Python Programming"
    original_fee = 5000
elif course_choice == 2:
    course_name = "Data Analytics"
    original_fee = 8000
elif course_choice == 3:
    course_name = "AI & ML"
    original_fee = 12000
else:
    print("Invalid course selection")
    exit()

# Calculate discount
discount = 0

if is_student == "yes":
    discount += original_fee * 0.10

if early_registration == "yes":
    discount += original_fee * 0.05


final_amount = original_fee - discount
print("\n--- Fee Details ---")
print("Course Name:", course_name)
print("Original Fee: ₹", original_fee)
print("Total Discount: ₹", discount)
print("Final Payable Amount: ₹", final_amount)
```
